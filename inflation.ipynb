```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
inflation = pd.read_excel('/home/brandonealr/Desktop/inflacion.xls')
inflation = inflation.dropna().reset_index(drop=True)
inflation

```

```python
inflation = inflation.drop('Unnamed: 3', axis=1)
inflation = inflation.drop('Unnamed: 4', axis=1)
inflation = inflation.drop('Unnamed: 6', axis=1)
inflation = inflation.drop('Unnamed: 7', axis=1)
inflation
```

```python
inflation.rename(columns={'Instituto Nacional de Estadística y Geografía (INEGI)': 'Periodos'}, inplace=True)
inflation.rename(columns={'Unnamed: 1': 'Inflación total'}, inplace=True)
inflation.rename(columns={'Unnamed: 2': 'Inflación subyacente'}, inplace=True)
inflation.rename(columns={'Unnamed: 5': 'Inflación no subyacente'}, inplace=True)
```

```python
inflation = inflation.drop(0)
inflation
```

```python
inflation['Periodos'].astype(str)
inflation.set_index('Periodos', inplace=True)
inflation.plot()
plt.xlabel('Periodos')
plt.ylabel('Tasa de Inflación')
plt.title('Tasa de inflación mensual interanual  México(2011-2022)')
plt.xticks(rotation=90)
plt.show()
```
